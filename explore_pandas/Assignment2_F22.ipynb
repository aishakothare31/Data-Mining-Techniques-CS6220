## Scatter plots

```python
iris_data['sepal_ratio']=iris_data['SepalLength_cm']/iris_data['SepalWidth_cm']
iris_data['petal_ratio']=iris_data['PetalLength_cm']/iris_data['PetalWidth_cm']
iris_data['sepal_petal_length']=iris_data['SepalLength_cm']/iris_data['PetalLength_cm']
iris_data['sepal_petal_width']=iris_data['SepalWidth_cm']/iris_data['PetalWidth_cm']
iris_data['sepal_petal_len:w']=iris_data['SepalLength_cm']/iris_data['PetalWidth_cm']
iris_data['sepal_petal_w:len']=iris_data['SepalWidth_cm']/iris_data['PetalLength_cm']
iris_data['id']=iris_data.index

sea.lmplot(x='id', y="sepal_ratio", data=iris_data, hue="Specie", fit_reg=False, legend=True)
sea.lmplot(x='id', y="petal_ratio", data=iris_data, hue="Specie", fit_reg=False, legend=True)
sea.lmplot(x='id', y="sepal_petal_length", data=iris_data, hue="Specie", fit_reg=False, legend=True)
sea.lmplot(x='id', y="sepal_petal_width", data=iris_data, hue="Specie", fit_reg=False, legend=True)
sea.lmplot(x='id', y="sepal_petal_len:w", data=iris_data, hue="Specie", fit_reg=False, legend=True)
sea.lmplot(x='id', y="sepal_petal_w:len", data=iris_data, hue="Specie", fit_reg=False, legend=True)
```
